```python
select_category = pn.widgets.Select(name='Select Category', options=['All','Exchange Rate Measures', 'Goods', 'Services', 'Capital Account',
       'Financial Sector', 'Applies to All', 'Resident', 'Non-Resident',
       'Payments_Inwards', 'Payments_Outwards', 'Trade_Inwards',
       'Trade_Outwards'])
select_category
```

```python
##### define plot function
def plot_expenses(category):
    if category == 'All':
        plot_df = df.groupby('Survey _ID')['ECLI_Index'].mean()
    else:
        plot_df = df[df['Category'] == category].groupby('Survey _ID')['ECLI_Index'].mean()
    plot = plot_df.hvplot.bar(stacked = True, legend="top_left",x='Survey _ID', y='ECLI_Index', title = "Dynamic ecli by category")
    return plot

# define callback function
@pn.depends(select_category.param.value)
def update_plot(category):
    plot = plot_expenses(category)
    return plot

# create layout
df_chart = pn.Row(select_category, update_plot)
df_chart[1].width = 600

df_chart
```
